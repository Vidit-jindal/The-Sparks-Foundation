**City wise Sales analysis**

```
df1 = df.groupby(['City'])[['Sales', 'Discount','Profit']].mean()
df1
```

```
df2 =  df['City'].value_counts().to_frame()
df1 = pd.merge(df1, df2, left_index=True, right_index=True)
df1
```

```
df1.rename(columns={'City':'NoOfDeals'},inplace = True)
```

```
#Sorted on basis of Profit in Dollars. Plotted the 50 Cities with Best Sales
df1.sort_values('Profit').tail(50)[['NoOfDeals','Sales','Profit']].plot(kind='bar',figsize= (20,10),label=['NoOfDeals','Avg Sales Price($)','Profit($)'])
plt.ylabel('No Of Deals/Average Sales Price($) and Profit per City($)')
plt.xlabel('Cities')
plt.legend()
plt.show()
```

```
#Sorted on basis of Profit in Dollars. Plotted the 50 Cities with least Sales
df1.sort_values('Profit').head(50)[['NoOfDeals','Sales','Profit']].plot(kind='bar',figsize= (20,10),label=['NoOfDeals','Avg Sales Price($)','Profit($)'])
plt.ylabel('No Of Deals/Average Sales Price($) and Profit per City($)')
plt.xlabel('Cities')
plt.legend()
plt.show()
```

```
#Sorted on basis of Profit.Plotted the 50 Cities with Best Profit
df1.sort_values('Profit')[['Discount','Profit']].tail(50).plot(kind='bar', figsize= (20,10))
plt.ylabel('Discount($) and Profit($) per City')
plt.xlabel('Cities')
plt.legend()
plt.show()
```

```
#Sorted on basis of Profit.Plotted the 50 Cities with Least Profit
df1.sort_values('Profit')[['Discount','Profit']].head(50).plot(kind='bar', figsize= (20,10))
plt.ylabel('Discount($) and Profit($) per City')
plt.xlabel('Cities')
plt.legend()
plt.show()
```
